(Extração das tabelas)

```python
import boto3
import pandas as pd
from functools import reduce
```

```python
s3 = boto3.client("s3", region_name="us-east-2")

arquivos = [
    "atendimento_sinistros.csv",
    "cadastro_clientes.csv",
    "churn_.csv",
    "contratos_apolices.csv",
    "engajamento_marketing.csv",
]

for arquivo in arquivos:
    s3.download_file(
        Bucket   = "t2rpt",
        Key      = arquivo,
        Filename = arquivo  # nome do arquivo que vai ser salvo localmente
    )
    print(f"✅ {arquivo} baixado!")
```
